### set unset proxy in WSL

I have just added some bash commands at the end of my `.bashrc` file.

```bash
# Set Proxy
function setproxy() {
    export {http,https,ftp}_proxy="http://<my proxy ip address>:80"
    export {HTTP,HTTPS,FTP}_PROXY="http://<my proxy ip address>:80"
}

# Unset Proxy
function unsetproxy() {
    unset {http,https,ftp}_proxy
    unset {HTTP,HTTPS,FTP}_PROXY
}
```
